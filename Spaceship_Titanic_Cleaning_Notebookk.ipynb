```python
import pandas as pd 
```

```python
# Reading Data
df = pd.read_csv(r"C:\Users\ELTANANY 01062856027\Downloads\Spaceship Titanic (2).csv")
df.head()
```

```python
# How To Discover Date
df.info()
```

```python
bool_cols = ['CryoSleep', 'VIP'] # Convert Data Type into boolean [TRUE,FALSE]

for col in bool_cols:
    if col in df.columns:
        df[col] = df[col].astype('boolean')
df.info()
```

# Handle Missing Values

```python
#for Spend Money 
spending_cols = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
for col in spending_cols:
    df.loc[(df['CryoSleep'] == True) & (df[col].isna()), col] = 0 # 'CryoSleep' == True >> always Sleep mode 

df[spending_cols].isna().sum()
```

```python
# set 0 for other missing values
for col in spending_cols:
    df[col] = df[col].fillna(0)
    
df[spending_cols].isna().sum()
```

```python
#For Age
df['Age'].isna().sum()
```

```python
# Set age accorrding to ['HomePlanet', 'VIP'] otherwise by median 
group_age_median = df.groupby(['HomePlanet', 'VIP'])['Age'].transform('median')
df['Age'] = df['Age'].fillna(group_age_median)

df['Age'] = df['Age'].fillna(df['Age'].median())

df['Age'].isna().sum()
```

```python
# calculate age group 
def make_age_group(age):
    if age <10:
        return 'Kids'
    if age < 13:
        return 'below_teen'
    elif age < 18:
        return 'teen'
    elif age < 35:
        return 'young_adult'
    elif age < 45:
        return 'adult'
    elif age < 60:
        return 'middle_age'
    else:
        return 'senior'

df['AgeGroup'] = df['Age'].apply(make_age_group)
df[['Age', 'AgeGroup']].head()
```

```python
#For Name
df.Name.isna().sum()
```

```python
#Replace na with unkmown
if 'Name' in df.columns:
    df['Name'] = df['Name'].fillna('Unknown')

df.Name.isna().sum()
```

```python
#for Cryosleep 
df.CryoSleep.isna().sum()
```

```python
#Totalspent==0 >>Cryosleep is True
df.loc[df['CryoSleep'].isna() & (df['TotalSpent'] == 0), 'CryoSleep'] = True
df.loc[df['CryoSleep'].isna() & (df['TotalSpent'] > 0), 'CryoSleep'] = False

df['CryoSleep'] = df['CryoSleep'].astype('boolean')
df['CryoSleep'].isna().sum()
```

```python
df.info()
```

```python
# for VIP 
df.VIP.isna().sum() 
```

```python
# majority is false
if 'VIP' in df.columns:
    df['VIP'] = df['VIP'].fillna(False)

df.VIP.isna().sum()
```

```python
#for destination 
df.Destination.isna().sum()
```

```python
# replace with unknown
for des in df.columns:
    df.Destination=df.Destination.fillna("Unknown")

df.Destination.isna().sum()
```

```python
#drop Na rows in ['Cabin', 'HomePlanet'] option 1 
#df = df.dropna(subset=['Cabin', 'HomePlanet'])
```

```python
df.info()
```

```python
#Option 2 
df['HomePlanet'] = df['HomePlanet'].replace('Unknown', pd.NA)

df['HomePlanet'] = df['HomePlanet'].fillna(df['HomePlanet'].mode()[0])
df.HomePlanet.isna().sum()
```

```python
#split cabin col
df[['Deck','CabinNum','Side']] = df['Cabin'].str.split('/', expand=True)
```

```python
for col in ['Deck', 'Side']:
    # mode >> HomePlanet
    mode_by_home = df.groupby('HomePlanet')[col].agg(
        lambda x: x.mode().iloc[0] if not x.mode().empty else pd.NA
    )
    
    df[col] = df[col].fillna(df['HomePlanet'].map(mode_by_home))
    
    # Otherwise backup mode 
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df['CabinNum'] = pd.to_numeric(df['CabinNum'], errors='coerce')

# median >> Deck
median_by_deck = df.groupby('Deck')['CabinNum'].transform('median')

df['CabinNum'] = df['CabinNum'].fillna(median_by_deck)

# backup median 
df['CabinNum'] = df['CabinNum'].fillna(df['CabinNum'].median())

#  int
df['CabinNum'] = df['CabinNum'].round().astype('Int64')
```

```python
#combination
df['Cabin'] = df['Deck'].astype(str) + '/' + df['CabinNum'].astype(str) + '/' + df['Side'].astype(str)
df.Cabin.isna().sum()
```

```python
path = r"C:\Users\ELTANANY 01062856027\Downloads\Spaceship Titanic (2).csv"

df.to_csv(path, index=False)
print(f'Cleaned file saved as: {path}')
```
